```python
# Importing the required libraries
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
# URL of the website
url = 'http://fondationscp.wikidot.com/scp-000'
```

```python
# Sending a request to the website
response = requests.get(url)
```

```python
# Parsing the HTML content
soup = bs(response.content, 'html.parser')
```

```python
# Extracting the title of the website
title = soup.title.get_text()
```

```python
# Extracting the text of the website
text = soup.get_text()
```

```python
# Printing the title and text of the website
print(title)
print(text)
```

```python
# Extracting all the links in the website
links = soup.find_all('a')

# Printing all the links in the website
for link in links:
    print(link.get('href'))
```

```python
# Print the Parse Tree of the website
print(soup.prettify())
```

```python
# Extracting all the paragraphs in the website
paragraphs = soup.find_all('p')

# Printing all the paragraphs in the website
for paragraph in paragraphs:
    print(paragraph.get_text())
```

```python
# Extracting all the headings in the website
headings = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

# Printing all the headings in the website
for heading in headings:
    print(heading.get_text())
```

```python
# Extracting all the images in the website
images = soup.find_all('img')

# Printing all the images in the website
for image in images:
    print(image.get('src'))
```

```python
# Creating a DataFrame from the extracted data
paragraphs_df = [p.text for p in paragraphs]
links_df = [a.get('href') for a in links]
images_df = [img.get('src') for img in images]

# Determine the maximum length
max_len = max(len(paragraphs_df), len(links_df), len(images_df))

# Extend the lists to the maximum length
paragraphs_df.extend([None]*(max_len - len(paragraphs_df)))
links_df.extend([None]*(max_len - len(links_df)))
images_df.extend([None]*(max_len - len(images_df)))

# Creating a DataFrame from the extracted data
data = {'Paragraphs': paragraphs_df, 'Links': links_df, 'Images': images_df}
df = pd.DataFrame(data)

print(df)
```

```python
# Saving the DataFrame to a CSV file
df.to_csv('data.csv', index=False)
```
